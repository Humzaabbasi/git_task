```python
#import libraries
import seaborn as sns
import numpy as np
from numpy import median
import matplotlib.pyplot as plt
#import data_set
boat=sns.load_dataset("titanic")
plt.figure(figsize=(4,4))
#plot the dataset
sns.barplot(x="class",y="fare",hue="sex",data=boat,order=["female","male"],color="blue",
           palette="flare",estimator=median)
plt.title("Titanic Barplot")
plt.show()

```

## Mean

```python
#import libraries
import seaborn as sns
from numpy import  mean
import matplotlib.pyplot as plt
#import data_set
boat=sns.load_dataset("titanic")
plt.figure(figsize=(4,4))
#plot the dataset
sns.barplot(x="class",y="fare",hue="sex",data=boat,estimator=mean,saturation=3)
plt.title("Titanic Barplot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean
phool=sns.load_dataset("iris")
sns.color_palette("hls", 8)
plt.figure(figsize=(6,4))
#phool
sns.barplot(x="sepal_width",y="petal_width",hue="species",data=phool,estimator=mean,
                ci=None,saturation=0.6)
plt.title("Iris Barplot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median
phool=sns.load_dataset("iris")
sns.color_palette("hls", 8)
plt.figure(figsize=(6,4))
#phool
sns.lineplot(x="sepal_width",y="petal_width",hue="species",data=phool,estimator=median,
                ci=None)
plt.title("Median")
plt.show()
```
