```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
sns.set_style(style=("whitegrid"))
kashti=sns.load_dataset("titanic")
kashti.head()
plt.title("box plot")
sns.boxplot(x="survived",y="age",data=kashti,saturation=0.9)
plt.show()
```

```python
p1=sns.set_style(style=("whitegrid"))
kashti=sns.load_dataset("titanic")
kashti.head()
plt.title("box plot")
sns.boxplot(x="survived",y="age",data=kashti,
            showmeans=True,saturation=1)
plt.show()
```

```python
sns.boxplot(x="survived",y="age",data=kashti,
            showmeans=True,saturation=1,
           meanprops={"marker":"+","markersize":"10","markeredgecolor":"red"})
#show labels
plt.xlabel("survived",size=14)
plt.ylabel("Age",size=14)
plt.title("box plot",size=15,weight="bold")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
sns.set_style(style=("whitegrid"))
boat=sns.load_dataset("titanic")
sns.boxplot(x="survived",y="age",data=boat,saturation=1,showmeans=True,
           meanprops={"marker":"+","markersize":"10","markeredgecolor":"red"})
plt.xlabel("male or female",size=10)
plt.ylabel("fare",size=10)
plt.title("box plot",size=15,weight="bold")
plt.show()
```

```python
# importing packages
import seaborn
import matplotlib.pyplot as plt
  
# loading of a dataframe from seaborn
df = seaborn.load_dataset('tips')
  
############# Main Section         #############
# Form a facetgrid using columns with a hue
graph = seaborn.FacetGrid(df, col ='time', hue ='smoker')
# map the above form facetgrid with some attributes
graph.map(seaborn.regplot, "total_bill", "tip").add_legend()
# show the object
plt.show()
  
# This code is contributed by Deepanshu Rustagi.
```
