```python
## - Adding X-Limits and Y-Limits
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#Load_data_set
flower=sns.load_dataset("iris")
#flower
#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=flower)
plt.title("set X and Y limits")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

```python
# # set styles
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#Load_data_set
flower=sns.load_dataset("iris")
plt.figure(figsize=(8,6))
#flower
#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=flower)
plt.title("set X and Y limits")
#sns.set(style="ticks", context="talk")
sns.set_style("dark")
sns.set_style(style=None,rc=None)
plt.show()
```
